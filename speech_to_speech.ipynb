```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
! pip install sentencepiece
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")

```

```python
text=" When my brother and I were starting our first company, instead of getting an apartment, we just rented a small office and we slept on the couch. And we showered at the YMCA and we're so hot up we had just one computer, so the website was up during the day and I was coding at night. Seven days a week all the time. If you do simple math, say like, okay, if somebody else is working 50 hours and you're working 100, you'll get twice as done as my brother."
```

```python
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")
model_inputs = tokenizer(text, return_tensors="pt")
```

```python

# translate from English to Hindi
generated_tokens = model.generate(
    **model_inputs,
    forced_bos_token_id=tokenizer.lang_code_to_id["hi_IN"]
)

```

```python
translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
translation
```

```python
import nltk
nltk.download('punkt')
sentences = nltk.sent_tokenize(text)
```

```python
import numpy as np
SPEAKER = "v2/hi_speaker_6"
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio
```

```python

pieces = []
for sentence in sentences:
    audio_array = generate_audio(sentence, history_prompt=SPEAKER)
    pieces += [audio_array]
```

```python
Audio(np.concatenate(pieces), rate=SAMPLE_RATE)
```
